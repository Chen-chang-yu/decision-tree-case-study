```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
# 當複雜係數 α = 0 時, 可獲得最大的葉節點數, 畫出決策樹
iris = load_iris()
clf = tree.DecisionTreeClassifier(ccp_alpha= 0.00, random_state=0)
clf = clf.fit(iris.data, iris.target)
tree.plot_tree(clf)
```

```python
# 設定圖片大小, 讓圖片可清楚顯示
fig = plt.figure(figsize = (10, 8), dpi = 150)
tree.plot_tree(clf)
```

```python
# 決策樹上色.
fig = plt.figure(figsize = (10, 8), dpi = 150)
tree.plot_tree(clf, filled = True)
```
